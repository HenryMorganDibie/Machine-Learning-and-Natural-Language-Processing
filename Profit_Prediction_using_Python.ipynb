```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("drive/MyDrive/Start ups/50_Startups.csv")
print(data.head())
```

```python
print(data.describe())
```

```python
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
x = data[["R&D Spend", "Administration", "Marketing Spend"]]
y = data["Profit"]

x = x.to_numpy()
y = y.to_numpy()
y = y.reshape(-1, 1)

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)
data = pd.DataFrame(data={"Predicted Profit": ypred.flatten()})
print(data.head())
```

So, using machine learning algorithms, we can forecast a company's profit for a given period. Such tasks can assist a company in establishing a realistic target.
